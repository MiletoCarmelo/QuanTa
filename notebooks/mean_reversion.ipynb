```python
from quanta.clients.yfinance import YahooFinanceClient
from datetime import datetime, timedelta

n=365
from_date = datetime.now() - timedelta(days=n)
to_date = datetime.now() - timedelta(days=1)

yh = YahooFinanceClient()
df_init = yh.get_price(
    "USO",
    from_date=from_date.strftime("%Y-%m-%d"),
    to_date=to_date.strftime("%Y-%m-%d"),
    interval="1h", 
    postclean=True
)

df_init.to_pandas()
```

```python
from quanta.utils.ta import TAClient, SMA, BollingerBands, ZScore
indicators = [
    SMA(period=20),
    BollingerBands(period=20, std_dev=2),
    ZScore(period=20)
]
```

```python
ta_client = TAClient()
df = ta_client.calculate_indicators(df_init, indicators)
df.columns  
```

```python
from quanta.clients.chart import ChartClient
chart_client = ChartClient()
chart_client.plot(df, "USO", indicators=indicators, theme='professional')
```
